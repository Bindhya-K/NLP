```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
#Step1. initialize sentences
sentences=['the glass of milk','cup of tea','I am studying','Veda is sleeping','My parents had returned today morning',
           'Satheesh is working','She has exams startinf tomorrow']
```

```python
# step 2 :vocab size
voc_size=10000
```

```python
#step2. One Hot Representation
onehot_rep = [one_hot(words,voc_size)for words in sentences]
```

```python
onehot_rep
```

```python
#Step 4: Word Embedding representation(no. of features ahs to be given)
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np

```

```python
# since each length are of different sizes, we need to make it equal lenth using pad sequence.  let here the sentence length be 8
sent_length = 8
pre_embedding = pad_sequences(onehot_rep,padding='pre',maxlen=sent_length)
pre_embedding
```

```python
# let the no. of feature be 15
dim=15
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length = sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(pre_embedding)
```
